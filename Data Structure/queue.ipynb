```python
class Queue:
    def __init__(self, N):
        self.q_list = [0 for i in range(N)]
        self.head = -1
        self.tail = -1
        
    def empty(self):
        if self.head > self.tail:
            return 1
        return 0
    
    def size(self):
        if self.tail <0:
            return 0
        return (self.tail - self.head + 1)
        
    def push(self, x):
        if self.head == -1:
            self.head += 1
        self.tail += 1
        self.q_list[self.tail] = x
    
    def pop(self):
        if self.empty() == 1:
            return -1
        else:
            self.head += 1
            return self.q_list[self.head-1]
        
    def front(self):
        if self.empty() == 1:
            return -1
        else:
            return self.q_list[self.head]
        
    def back(self):
        if self.empty() == 1:
            return -1
        else:
            return self.q_list[self.tail]
        
        
N = int(input())
q = Queue(N)

for _ in range(N):
    
    act = input().split()
    
    if act[0] == 'push':
        num = int(act[-1])
        q.push(num)
    elif 'pop' in act:
        print(q.pop())
    elif 'empty' in act:
        print(q.empty())
    elif 'front' in act:
        print(q.front())
    elif 'back' in act:
        print(q.back())
    elif 'size' in act:
        print(q.size())
```
